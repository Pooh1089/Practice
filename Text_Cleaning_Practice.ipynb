### 3. Remove URL

```python
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+') # \S AND \s are different
    return pattern.sub("",text) # sub means substitute
    #print(text)
```

```python
text1 = "check out youtube channel https://www.youtube.com/watch?v=V9tJCQoBakA&t=225s"
text2 = " Google www.google.com"
text3 = " remove https://dummy.com"
remove_url(text1)
```

```python
df['text'] = df['text'].apply(remove_url)
df['text'][:50]
```

### 4. Remove Punctuation

```python
import string
```

```python
exclude = string.punctuation
```

```python
#Remove punctuation
def remove_punc(text):
    for char in exclude :
        text = text.replace(char,"")
    return text
```

```python
#Method is faster compared to previous remove_punc
def remove_punc1(text):
    return text.translate(str.maketrans("","",exclude))
```

```python
text = " string@# () . !"
remove_punc(text)
```

```python
remove_punc1(text)
```

```python
df['text'] = df['text'].apply(remove_punc1)
df['text'][:50]
```

```python
chat_words_map_dict = {}
chat_words_list = []
for line in chat_words_str.split("\n"):
    if line != "":
        cw = line.split("=")[0]
        cw_expanded = line.split("=")[1]
        chat_words_list.append(cw)
        chat_words_map_dict[cw] = cw_expanded
chat_words_list = set(chat_words_list)

def chat_words_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words_list:
            new_text.append(chat_words_map_dict[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)

chat_words_conversion("one minute BRB")
```

### 6. Remove emojis

```python
def remove_emoji(string):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)

remove_emoji("game is on 🔥🔥")
```

```python
df['text'] = df['text'].apply(remove_emoji)
df['text'][:50]
```

### Remove Stop words

```python
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download("stopwords")
```

```python
stopwords.words("english")
```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])
```

```python
text = " I am Pooja. I am learning NLP"
remove_stopwords(text)
```

```python
df["text"] = df["text"].apply(lambda text: remove_stopwords(text))
df.head()
```

## Stemming

```python
from nltk.stem import PorterStemmer
```

5 Advantages and 5 Disdvantages of BOW model:

Advantages
1. Simplicity: BoW is straightforward to understand and implement. It doesn't require complex linguistic or semantic knowledge, making it accessible for beginners in natural language processing (NLP).

2. Efficiency: Processing text into a BoW representation is computationally efficient, allowing for quick feature extraction which is essential for large datasets.

3. Flexibility: BoW models can be easily scaled and adapted to different datasets and problem types, from simple classification tasks to more complex natural language understanding challenges.

4. Compatibility: The BoW representation works well with various machine learning algorithms, especially those designed for classification, such as Naive Bayes, Logistic Regression, and Linear SVMs.

5. Good Performance in Certain Contexts: For tasks where the presence of specific words is strongly indicative of the output, such as spam detection or certain types of sentiment analysis, BoW models can perform surprisingly well despite their simplicity.

Disadvantages
1. Loss of Context: BoW models ignore the order of words, which means the context and the meaning conveyed by the order are lost. This can lead to a significant decrease in the model's ability to understand the nuances of language.

2. Sparsity: The BoW representation can become very sparse, especially with large vocabularies, leading to high-dimensional but mostly empty feature vectors. This sparsity can be problematic for many machine learning algorithms and can lead to inefficient use of memory and computation.

3. Vocabulary Size: The size of the vocabulary can become very large, especially with datasets containing diverse topics or informal language with misspellings and slang. This can increase the computational resources required for training and inference.

4. Ambiguity and Polysemy: Words with multiple meanings (polysemy) are treated the same as their context is not considered. Similarly, synonyms are treated as distinct entities, potentially diluting important semantic similarities between texts.

5. Ignoring Semantics: BoW models do not capture the semantics or the meaning of the words. Advanced NLP tasks, which require understanding of the text, such as question answering or text summarization, cannot be effectively solved with BoW models alone.

5 Advantages and 5 Disdvantages of 1 hot encoding model:

Advantages
1. Simplicity: One-hot encoding is a straightforward and intuitive method to transform categorical data into a machine-readable form. It is easy to implement and understand, making it accessible for beginners in data science.

2. Compatibility with Machine Learning Models: Many machine learning algorithms require numerical input and do not handle categorical data well. One-hot encoding converts categorical data into a numerical format, making it possible to feed this data into almost any machine learning model.

3. No Implicit Ordering: One-hot encoding treats each category as equal without implying any sort of order (unless it is explicitly introduced). This is particularly useful for nominal categories where no ordinal relationship exists.

4. Improves Model Performance: For certain models, especially linear models, one-hot encoding can lead to an improvement in model performance as it allows the model to exploit the information in categorical variables correctly.

5. Clear Interpretation: The resulting data structure from one-hot encoding is clear and easy to interpret. Each bit represents a clear presence or absence of a category, which can simplify the analysis and understanding of the model's decisions.

Disadvantages
1. Dimensionality Increase: One-hot encoding can significantly increase the dataset's dimensionality (number of features), especially if the categorical variable has many unique values. This can lead to the "curse of dimensionality", which negatively impacts model performance and increases computational complexity.

2. Sparse Matrix: The resulting data from one-hot encoding is often sparse because for each record, only one of the one-hot encoded columns contains a 1, with the rest being 0. This sparsity can lead to inefficient memory usage and computational performance issues.

3. Loss of Information: One-hot encoding does not capture any information about the relationships between categories since it treats all categories as independent. Any ordinal relationship or ability to measure similarity between categories is lost.

4. Does Not Scale Well: For variables with a large number of categories, one-hot encoding can lead to a massive increase in the dataset's dimensionality, making models much harder to train and requiring significantly more computational resources.

5. Model Specificity: While one-hot encoding is beneficial for linear models, it may not be necessary or could even be detrimental for tree-based models. Tree-based models can handle categorical variables naturally and may perform better without one-hot encoding, as it preserves the ability to split on the categorical variable directly.



5 Advantages and 5 Disdvantages of TFIDF model:

Advantages
1. Relevance Measurement: TF-IDF provides a numerical statistic that reflects how important a word is to a document in a collection. This is useful for extracting relevant terms in documents, aiding in search, information retrieval, and data mining tasks.

2. Dimensionality Reduction: By prioritizing words that are unique to documents and penalizing common words across the corpus, TF-IDF can help reduce the dimensionality of the dataset, focusing on more meaningful features for machine learning models.

3. Improvement Over Simple Counts: Unlike simple bag-of-words models that only count terms, TF-IDF accounts for both the frequency of terms within each document and their inverse frequency across all documents, offering a more nuanced feature set for model training.

4. Automatic Keyword Extraction: TF-IDF is effective for automatic keyword extraction from documents, as terms with high TF-IDF scores are often more relevant to the document's subject matter, making it useful for summarization and topic modeling.

5. Versatility: The TF-IDF model can be easily integrated with various machine learning algorithms, enhancing the performance of models in tasks like classification, clustering, and search by providing them with more informative feature sets.

Disadvantages
1. Loss of Order Information: TF-IDF, much like bag-of-words models, ignores the order of terms within the document. This means contextual nuances and the semantics of word order are lost, which can be crucial for understanding the meaning of texts.

2. Ignores Document Length: TF-IDF does not necessarily account for the varying lengths of documents; longer documents may have inflated term frequencies simply due to their length, not because those terms are more relevant.

3. Not Suitable for Semantic Similarity: Since TF-IDF focuses on term frequencies, it is not designed to capture semantic relationships between words or phrases. Words with similar meanings but different lexical forms are treated as completely distinct features.

4. Requires a Comprehensive Corpus: The effectiveness of the inverse document frequency component depends on having a large and representative corpus. If the corpus is too small or not diverse enough, the IDF scores may not accurately reflect the importance of terms across different contexts.

5. Hyperparameter Sensitivity: The performance of TF-IDF can be sensitive to the choice of hyperparameters, such as the formula for computing term and document frequencies, and the method for normalizing these frequencies. Finding the optimal settings can require extensive experimentation and tuning.
